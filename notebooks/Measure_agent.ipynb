```python

from connect4.agent.random_agent import RandomAgent
from connect4.agent.min_max import MinMaxAgentWAlphaBeta
from connect4.agent.monte_carlo import MonteCarlo
from connect4.board import Board, PLAYER1, PLAYER2
from connect4.evaluator import AdvancedScore

from matplotlib import pyplot as plt
from time import time
```

```python
def measure(player1):
    player2 =RandomAgent
    player1.player = PLAYER1
    rounds = 100
    times= []
    length = []
    
    for _ in range(rounds):
        board = Board()
        player = PLAYER1
        t=[]
    
        while not board.is_game_over():
            if player == PLAYER1:
                start_time = time()
                col = player1.move(board)
                t.append(time()-start_time)
                board.add_token(col)
                player = PLAYER2
            elif player == PLAYER2:
                col = player2.move(board)
                board.add_token(col)
                player = PLAYER1
            print('.', end='')
        print()
        times.append(sum(t)/len(t))
        length.append(board.move_count)
    print('done')
    return sum(times)/len(times), sum(length)/len(length)
```

```python
print(measure(RandomAgent))
```

```python
mont = []
mont.append(measure(MonteCarlo(100)))
mont.append(measure(MonteCarlo(500)))
mont.append(measure(MonteCarlo(1_000)))
mont.append(measure(MonteCarlo(1_500)))
mont.append(measure(MonteCarlo(3_000)))
mont.append(measure(MonteCarlo(10_000)))
print(mont)
```

```python
adv = AdvancedScore()
minmaxab=[]
minmaxab.append(measure(MinMaxAgentWAlphaBeta(3, adv.score)))
minmaxab.append(measure(MinMaxAgentWAlphaBeta(4, adv.score)))
minmaxab.append(measure(MinMaxAgentWAlphaBeta(5, adv.score)))
minmaxab.append(measure(MinMaxAgentWAlphaBeta(6, adv.score)))
minmaxab.append(measure(MinMaxAgentWAlphaBeta(7, adv.score)))
print(minmaxab)
```
